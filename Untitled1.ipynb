```python
import layoutparser as lp
import uuid
import os
import numpy as np
import pandas as pd


def layout_det(image_obj):
    model_tt = lp.Detectron2LayoutModel('lp://PubLayNet/mask_rcnn_X_101_32x8d_FPN_3x/config',extra_config=["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.65],label_map={0: "Text", 1: "Title", 2: "List", 3:"Table", 4:"Figure"})
    data=[]
    for page,image in enumerate(image_obj):
        im_np = np.asarray(image)
        im_np = im_np[..., ::-1]
        layout = model_tt.detect(im_np)
        df=layout.to_dataframe()
        df["page_index"] = page
        df=df.sort_values(by=['y_1'], ascending=True)
        data.append(df)
    csv=pd.concat([x for x in data])
    csv.to_csv(f"{path}/layout.csv",index=False)
    
def ocr_image(filename):
    pdf_layout, pdf_images = lp.load_pdf(filename, load_images=True)
    return pdf_layout,pdf_images

def image_save(image_obj):
    for page,image in enumerate(image_obj):
        image.save(f"{path}/image/{page}.png","PNG")
    print("image done")
        
def ocr_df(ocr_obj):
    data=[]
    for page,df in enumerate(ocr_obj):
        df=df.to_dataframe()
        df["page_index"] = page
        data.append(df)
    csv=pd.concat([x for x in data])
    csv.to_csv(f"{path}/ocr.csv",index=False)
    print("Done csv")
    
def process(pdf):
    pdf_layout,pdf_images=ocr_image(pdf)
    image_save(pdf_images)
    layout_det(pdf_images)
    ocr_df(pdf_layout)

path=f"data/{uuid.uuid4().hex}"
os.makedirs(f"{path}/image",exist_ok=True)
process("test.pdf")
```

```python
def get_center(x1,y1,x2,y2):
    xCenter = (x1 + x2) / 2
    yCenter = (y1 + y2) / 2
    return xCenter,yCenter

def FindPoint(x1, y1, x2,
              y2, x, y) :
    if (x > x1 and x < x2 and
        y > y1 and y < y2) :
        return True
    else :
        return False
    
def check_point(model_bbx,word_bbx):
    point_x,point_y = get_center(word_bbx[0],word_bbx[1],
                                 word_bbx[2],word_bbx[3])
    return FindPoint(model_bbx[0],model_bbx[1],
                     model_bbx[2],model_bbx[3],point_x,point_y)

# 57.289883	72.284431	535.648438	108.862114
```

```python
data=[]

for idn,model_row in df.iterrows():
    tp_bbx=model_row[:4].astype(int).values
    para_data = {}
    word_index=[]
    ocr_test=ocr[ocr.page_index==model_row.page_index]
    
    for idx,ocr_row in ocr_test.iloc[:,:4].iterrows():
        word_bbx=ocr_row.astype(int).values
        flag=check_point(tp_bbx,word_bbx)
        if flag:
            word_index.append(idx)
    para_data["page_index"]=model_row.page_index
    para_data["type"]=model_row.type
    para_data["word_index"]=word_index
    
    data.append(para_data)


    
```
